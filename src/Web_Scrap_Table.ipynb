```python
def webSC(td):
    total=[]
    new_lst=[]
    inside_lst=[]
    count=0
    for ind,elem in enumerate(td):
        if len(new_lst) == mult[count]:
            #print(mult[count2])
            total.append(new_lst)
            new_lst=[elem.text.strip()]
            count+=1
        else:
            new_lst.append(elem.text.strip())

    last = [elem.text.strip() for elem in td[-12:]]
    total.append(last)
    
    return total


def giveMyTable(lista,n,categ,public,category,prod_category):
    dct={}
    lst=[]
    for ind,val in enumerate(lista):
        if ind == 0:
            elem = val
        elif ind%n == 0:
            dct[elem]=lst
            elem = val
            lst=[]
        else:
            lst.append(val)
    dct[elem]=lst
    dct["public"]=[public]*(n-1)
    dct["category"]=[category]*(n-1)
    dct["prod_category"]=[prod_category]*(n-1)
    dct["original"]=[categ]*(n-1)
    return dct
```
